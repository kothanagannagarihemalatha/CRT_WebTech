```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Now we need to convert some categorical variables into dummy variables.

```python
dataset2 = pd.get_dummies(dataset , columns=['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
```

```python
dataset2.head()
```

```python
cols=['cp_0','cp_1','cp_1','cp_2','cp_3','trestbps','chol','restecg_0','restecg_1','restecg_2','thalach']
```

```python
x=dataset2[cols]
y=dataset.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x , y , test_size=0.25 , random_state= 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
x_test
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test , y_pred)
sns.heatmap(cm , annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```

```python
import pickle
pickle.dump(classifier , open('model.pkl', 'wb'))
pickle.dump(sc, open('sc.pkl', 'wb'))
```
